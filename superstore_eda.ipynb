# üìä Superstore Sales Analysis - Exploratory Data Analysis

## üéØ Project Overview
This project performs comprehensive exploratory data analysis (EDA) on Superstore sales data to uncover business insights and identify opportunities for revenue growth.

## üìã Dataset Information
- **Source:** Superstore Sales Dataset
- **Records:** 9,994 transactions
- **Time Period:** 2014-2017
- **Columns:** 21 features including sales, profit, customer, and product information

## üéØ Business Questions
1. What are the sales trends over time?
2. Which product categories generate the most revenue?
3. Which regions perform best?
4. What are the most profitable products?
5. How do customer segments differ in purchasing behavior?
6. What is the shipping performance?

## üõ†Ô∏è Tools & Libraries
- Python 3.13.7
- Pandas for data manipulation
- Matplotlib & Seaborn for visualization
- Jupyter Notebook for interactive analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
# Settings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
```

```python
# Plotting style
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
```

```python
# Cell 2: Load dataset
df = pd.read_csv('Sample - Superstore.csv', encoding='latin-1')

print(f"Dataset loaded successfully!")
print(f"Shape: {df.shape}")
```

```python
# Cell 6: Statistical summary
df.describe()
```

```python
# Cell 7: Missing values
print("Missing values:")
print(df.isnull().sum())
```

```python
# Cell 8: Duplicates
print(f"Duplicate rows: {df.duplicated().sum()}")
```

```python
# Cell 9: Convert date columns to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df.dtypes
```

```python
df.head()
```

```python
# Cell 10: Create new useful columns

# 1. Extract year, month, day from Order Date
df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month
df['Order Month Name'] = df['Order Date'].dt.month_name()
df['Order Day'] = df['Order Date'].dt.day
df['Order Day Name'] = df['Order Date'].dt.day_name()

# 2. Calculate shipping time (days)
df['Shipping Days'] = (df['Ship Date'] - df['Order Date']).dt.days

# 3. Profit Margin (%)
df['Profit Margin'] = (df['Profit'] / df['Sales']) * 100

print("‚úÖ New features created!")
print(f"\nNew columns added:")
print(df[['Order Date', 'Order Year', 'Order Month Name', 'Shipping Days', 'Profit Margin']].head())
```

```python
# Cell 11: Validate cleaned data

print("=== CLEANED DATASET INFO ===")
print(f"Shape: {df.shape}")
print(f"\nData types:")
print(df.dtypes)
print(f"\nMissing values:")
print(df.isnull().sum().sum())
print(f"\nDate range:")
print(f"From: {df['Order Date'].min()}")
print(f"To: {df['Order Date'].max()}")
print("\n‚úÖ Data cleaning complete!")
```

```python
# Cell 12: Overall business metrics

print("=== BUSINESS OVERVIEW ===\n")

total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()
total_orders = df['Order ID'].nunique()
total_customers = df['Customer ID'].nunique()
avg_order_value = df.groupby('Order ID')['Sales'].sum().mean()
overall_profit_margin = (total_profit / total_sales) * 100

print(f"Total Sales: ${total_sales:,.2f}")
print(f"Total Profit: ${total_profit:,.2f}")
print(f"Total Orders: {total_orders:,}")
print(f"Total Customers: {total_customers:,}")
print(f"Average Order Value: ${avg_order_value:,.2f}")
print(f"Overall Profit Margin: {overall_profit_margin:.2f}%")
```

```python
# Cell 13: Sales by Year
sales_by_year = df.groupby('Order Year')['Sales'].sum().reset_index()
print("\n=== SALES BY YEAR ===")
print(sales_by_year)

# Visualization
plt.figure(figsize=(10, 6))
plt.bar(sales_by_year['Order Year'], sales_by_year['Sales'], color='skyblue', edgecolor='black')
plt.title('Total Sales by Year', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Sales ($)', fontsize=12)
plt.xticks(sales_by_year['Order Year'])
plt.grid(axis='y', alpha=0.3)

# Format y-axis untuk currency
ax = plt.gca()
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x/1000:.0f}K'))

plt.tight_layout()
plt.show()
```

```python
# Cell 14: Sales by Category
category_sales = df.groupby('Category').agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Quantity': 'sum',
    'Order ID': 'count'
}).round(2)

category_sales.columns = ['Total Sales', 'Total Profit', 'Units Sold', 'Orders']
category_sales['Profit Margin %'] = ((category_sales['Total Profit'] / category_sales['Total Sales']) * 100).round(2)
category_sales = category_sales.sort_values('Total Sales', ascending=False)

print("\n=== CATEGORY PERFORMANCE ===")
print(category_sales)

# Visualization
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Sales by Category
axes[0].barh(category_sales.index, category_sales['Total Sales'], color=['#FF6B6B', '#4ECDC4', '#45B7D1'])
axes[0].set_title('Total Sales by Category', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Sales ($)', fontsize=11)
axes[0].grid(axis='x', alpha=0.3)

# Profit by Category
axes[1].barh(category_sales.index, category_sales['Total Profit'], color=['#95E1D3', '#F38181', '#AA96DA'])
axes[1].set_title('Total Profit by Category', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Profit ($)', fontsize=11)
axes[1].grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Cell 15: Regional Performance
regional_perf = df.groupby('Region').agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Order ID': 'nunique',
    'Customer ID': 'nunique'
}).round(2)

regional_perf.columns = ['Total Sales', 'Total Profit', 'Orders', 'Customers']
regional_perf = regional_perf.sort_values('Total Sales', ascending=False)

print("\n=== REGIONAL PERFORMANCE ===")
print(regional_perf)

# Visualization - Pie Chart
plt.figure(figsize=(10, 7))
colors = ['#FF9999', '#66B2FF', '#99FF99', '#FFD700']
plt.pie(regional_perf['Total Sales'], 
        labels=regional_perf.index, 
        autopct='%1.1f%%',
        colors=colors,
        startangle=90,
        textprops={'fontsize': 12, 'fontweight': 'bold'})
plt.title('Sales Distribution by Region', fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()
```

```python
# Cell 16: Segment Analysis
segment_analysis = df.groupby('Segment').agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Customer ID': 'nunique',
    'Order ID': 'nunique'
}).round(2)

segment_analysis.columns = ['Total Sales', 'Total Profit', 'Customers', 'Orders']
segment_analysis['Avg Order Value'] = (segment_analysis['Total Sales'] / segment_analysis['Orders']).round(2)

print("\n=== CUSTOMER SEGMENT ANALYSIS ===")
print(segment_analysis)

# Visualization
fig, ax = plt.subplots(figsize=(10, 6))
x = np.arange(len(segment_analysis.index))
width = 0.35

bars1 = ax.bar(x - width/2, segment_analysis['Total Sales'], width, label='Sales', color='#3498db')
bars2 = ax.bar(x + width/2, segment_analysis['Total Profit'], width, label='Profit', color='#2ecc71')

ax.set_xlabel('Customer Segment', fontsize=12, fontweight='bold')
ax.set_ylabel('Amount ($)', fontsize=12, fontweight='bold')
ax.set_title('Sales & Profit by Customer Segment', fontsize=14, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(segment_analysis.index)
ax.legend()
ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Cell 17: Top 10 Products by Sales
top_products = df.groupby('Product Name').agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Quantity': 'sum'
}).round(2)

top_products = top_products.sort_values('Sales', ascending=False).head(10)

print("\n=== TOP 10 PRODUCTS BY SALES ===")
print(top_products)

# Visualization
plt.figure(figsize=(12, 7))
plt.barh(range(len(top_products)), top_products['Sales'], color='coral', edgecolor='black')
plt.yticks(range(len(top_products)), top_products.index, fontsize=9)
plt.xlabel('Sales ($)', fontsize=12, fontweight='bold')
plt.title('Top 10 Products by Sales', fontsize=14, fontweight='bold')
plt.gca().invert_yaxis()
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Cell 18: Shipping Performance
print("\n=== SHIPPING ANALYSIS ===")
print(f"Average Shipping Days: {df['Shipping Days'].mean():.2f} days")
print(f"Median Shipping Days: {df['Shipping Days'].median():.0f} days")
print(f"Fastest Shipping: {df['Shipping Days'].min()} days")
print(f"Slowest Shipping: {df['Shipping Days'].max()} days")

# By Ship Mode
shipping_mode = df.groupby('Ship Mode')['Shipping Days'].agg(['mean', 'count']).round(2)
shipping_mode.columns = ['Avg Days', 'Orders']
print("\nShipping by Mode:")
print(shipping_mode)

# Visualization
plt.figure(figsize=(10, 6))
plt.bar(shipping_mode.index, shipping_mode['Avg Days'], color='lightgreen', edgecolor='black')
plt.title('Average Shipping Days by Ship Mode', fontsize=14, fontweight='bold')
plt.xlabel('Ship Mode', fontsize=12)
plt.ylabel('Average Days', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Cell 19: Sales Trends by Month
monthly_sales = df.groupby('Order Month Name')['Sales'].sum().reindex([
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
])

print("\n=== MONTHLY SALES ===")
print(monthly_sales)

# Visualization
plt.figure(figsize=(14, 6))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o', linewidth=2, markersize=8, color='purple')
plt.fill_between(range(len(monthly_sales)), monthly_sales.values, alpha=0.3, color='purple')
plt.title('Sales Trend by Month', fontsize=16, fontweight='bold')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Sales ($)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Cell 20: Summary of Key Findings

print("=" * 60)
print("          KEY BUSINESS INSIGHTS & RECOMMENDATIONS")
print("=" * 60)

# 1. Category Insights
print("\nüìä CATEGORY PERFORMANCE:")
top_category = df.groupby('Category')['Sales'].sum().idxmax()
top_category_sales = df.groupby('Category')['Sales'].sum().max()
print(f"‚úì {top_category} is the top-performing category")
print(f"  Revenue: ${top_category_sales:,.2f}")

# Find least profitable category
least_profit_cat = df.groupby('Category')['Profit'].sum().idxmin()
print(f"‚úì {least_profit_cat} has the lowest profitability - needs review")

# 2. Regional Insights
print("\nüó∫Ô∏è REGIONAL PERFORMANCE:")
top_region = df.groupby('Region')['Sales'].sum().idxmax()
top_region_sales = df.groupby('Region')['Sales'].sum().max()
print(f"‚úì {top_region} region leads with ${top_region_sales:,.2f} in sales")

# 3. Customer Segment
print("\nüë• CUSTOMER SEGMENTS:")
top_segment = df.groupby('Segment')['Sales'].sum().idxmax()
segment_count = df.groupby('Segment')['Customer ID'].nunique()
print(f"‚úì {top_segment} segment generates highest revenue")
print(f"‚úì Customer distribution: {segment_count.to_dict()}")

# 4. Product Insights
print("\nüèÜ PRODUCT PERFORMANCE:")
top_product = df.groupby('Product Name')['Sales'].sum().idxmax()
top_product_sales = df.groupby('Product Name')['Sales'].sum().max()
print(f"‚úì Best-selling product: {top_product}")
print(f"  Total sales: ${top_product_sales:,.2f}")

# 5. Time-based Insights
print("\nüìÖ SALES TRENDS:")
best_month = df.groupby('Order Month Name')['Sales'].sum().idxmax()
best_month_sales = df.groupby('Order Month Name')['Sales'].sum().max()
print(f"‚úì Peak sales month: {best_month} (${best_month_sales:,.2f})")

best_year = df.groupby('Order Year')['Sales'].sum().idxmax()
yearly_growth = df.groupby('Order Year')['Sales'].sum()
if len(yearly_growth) > 1:
    growth_rate = ((yearly_growth.iloc[-1] - yearly_growth.iloc[0]) / yearly_growth.iloc[0]) * 100
    print(f"‚úì Overall growth rate: {growth_rate:.2f}%")

# 6. Shipping Performance
print("\nüì¶ SHIPPING INSIGHTS:")
avg_shipping = df['Shipping Days'].mean()
print(f"‚úì Average shipping time: {avg_shipping:.1f} days")
fastest_mode = df.groupby('Ship Mode')['Shipping Days'].mean().idxmin()
print(f"‚úì Fastest shipping mode: {fastest_mode}")

# 7. Profitability
print("\nüí∞ PROFITABILITY:")
overall_margin = (df['Profit'].sum() / df['Sales'].sum()) * 100
print(f"‚úì Overall profit margin: {overall_margin:.2f}%")

# Find most profitable sub-category
top_profit_subcat = df.groupby('Sub-Category')['Profit'].sum().idxmax()
top_profit_subcat_value = df.groupby('Sub-Category')['Profit'].sum().max()
print(f"‚úì Most profitable sub-category: {top_profit_subcat}")
print(f"  Profit: ${top_profit_subcat_value:,.2f}")

print("\n" + "=" * 60)
```
